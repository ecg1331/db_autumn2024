```python
# randomly generated by me
# sell by date
sd = datetime(2018, 3, 20)
ed = datetime(2024, 11, 30)

def gen_date():
    return sd + timedelta(days=random.randint(0, (ed- sd).days))
```

```python
sales_prefix = 90
random_sales_numbers =  f"{sales_prefix}{str(random.randint(100, 999)).zfill(4)}"

sales = []
for i in range(400):
    sales_id = f"{sales_prefix}{str(random.randint(100, 999)).zfill(4)}"
    employee_id = random.choice(employee_ids)
    customer_loyalty = str(random.choice(loyalty_ids)) if random.random() > 0.4 else None
    date = gen_date()

    sales.append({"Sale_ID": sales_id, "Employee_ID": employee_id, "Customer_Loyalty": customer_loyalty, "Date": date})

```

```python
sales_id = sales_df['Sale_ID'].to_list()
menu_item = menu['MenuItemSkew'].to_list()
drink = drinks['DrinkSkew'].to_list()
book_id = books['isbn'].to_list()
pastry = pastry['PastrySkew'].to_list()

possible_menu_items = menu_item + drink + pastry
```

```python
from random import choices

sales_item = []

for id in sales_id:

    num_items = random.randint(1, 6)
    items = random.sample(possible_menu_items, num_items)

    for item in items:
        quantity = choices([1, 2, 3, 4, 5], weights=[0.7, 0.2, 0.5, .25, .25], k=1)[0]
        sales_item.append({"Sale_ID": id,
                            "Item": item,
                            "Quantity": random.randint(1, 3)})
    
    num_books = random.randint(0, 3)
    books = random.sample(book_id, num_books)

    if len(books) > 0:
        for book in books:
            quantity = choices([1, 2, 3], weights=[0.7, 0.2, 0.1], k=1)[0]
            sales_item.append({"Sale_ID": id,
                                "Item": book,
                                "Quantity": quantity})
    
```
