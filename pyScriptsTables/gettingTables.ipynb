```python
breakfast = pd.read_csv("/Users/emma/Desktop/database/final/menu/breakfast_menu_plein_air_cafe.csv", index_col=False)
salads = pd.read_csv("/Users/emma/Desktop/database/final/menu/salads_sandwiches_baguettes_menu_plein_air_cafe.csv", index_col=False)
bowls = pd.read_csv("/Users/emma/Desktop/database/final/menu/tartines_plein_bowls_menu_plein_air_cafe.csv", index_col=False)
```

```python
full_menu = pd.concat([breakfast, salads, bowls]).reset_index(drop=True)
```

- need to lower the descri
    - take away anything thats not an asscii
- take away $ from price

```python
def clean_desc(txt):
    txt = txt.lower()
    txt = re.sub(r'[^a-z, ]', '', txt)
    return txt
```

```python
# cleaning menu
full_menu['Description'] = full_menu['Description'].apply(clean_desc)
full_menu['Price'] = full_menu['Price'].str.replace('$', '').astype(float)

full_menu
```

- Menu table: Id, category, price
- Ingredients table, ingreident id, ingredient name
- recipes table item id, recipe id

```python
# MENU ITEMS WILL BE IN THE 100s
menu_items_table = full_menu[['Item Name', 'Category', 'Price']].reset_index(drop=True)
menu_items_table['MenuItemSkew'] = menu_items_table.index + 100
menu_items_table = menu_items_table[['MenuItemSkew', 'Item Name', 'Category', 'Price']]
```

```python
menu_items_table.to_csv("/Users/emma/Desktop/database/final/tables/menu_items_table.csv", index=False)
menu_items_table.head()
```

```python
# merging bc its easier than appending
full_menu = full_menu.merge(menu_items_table[['MenuItemSkew', 'Item Name']], on='Item Name', how='left')
full_menu
```

```python
# Explode ingredients while keeping MenuItemID
ingredients_expanded = full_menu.assign(
    Ingredient = full_menu['Description'].str.split(r', | & | and | on |  | or', regex=True) # making sure it splits as much as possible
).explode('Ingredient').reset_index(drop=True)
ingredients_expanded.to_csv('ingredients_ex.csv')
```

```python
def add_ingredient(oldRow, ingredient):
    row = ingredients_expanded.loc[oldRow].copy()
    row['Ingredient'] = ingredient
    
    return pd.DataFrame([row])
```
