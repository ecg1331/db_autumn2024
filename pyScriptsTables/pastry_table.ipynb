```python
pastries = pd.concat([pastries, dietRes], ignore_index=True)
pastries
```

```python
pastries.drop(columns=['Quantity','Amount'], inplace=True)
```

```python
pastries.rename(columns={'Description': 'Pastry Name',
                         'Price Each' : 'Price'}, inplace=True)
```

```python
# PASTRIES WILL BE 4
pastries["PastrySkew"] = pastries.index + 400
```

```python
# sell by date
sd = datetime(2024, 10, 20)
ed = datetime(2024, 12, 31)

def gen_date():
    return sd + timedelta(days=random.randint(0, (ed- sd).days))
```

```python
pastries['Sell_by_Date'] = [gen_date() for x in range(len(pastries))]
```

```python
pastries = pastries[['PastrySkew', 'Pastry Name', 'Price', 'Sell_by_Date']]
```
